```python
#create a histogram of total bill amounts
sns.histplot(data=tips,x="total_bill")
```

```python
#Commonly used plot types:
#1. Line plot
#2. Scatter Plot 
#3. Histogram
#4. Box Plot
#5. Violin Plot
#6. Heatmap
#7. Pairplot
```

```python
#Scatter Plot
tips=sns.load_dataset("tips")
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
#Scatter Plot
import matplotlib.pyplot as plt 
tips=sns.load_dataset("tips")
sns.scatterplot(x="total_bill", size="size",hue="sex",sizes=(50,200),y="tip",data=tips)

plt.xlabel("Total Bill")
plt.ylabel("Tips")
plt.title("Scatter Plot of Relationship between Total Bill vs Tips")
plt.show()
```

```python
#line plot
fmri=sns.load_dataset("fmri")
fmri.tail(50)
```

```python
sns.lineplot(x="timepoint",y="signal",data=fmri)
```

```python
sns.lineplot(x="timepoint",hue="event",style="region",markers=True,dashes=False, y="signal",data=fmri)

#add labels and title
plt.title('Signal Over Time')
plt.xlabel("Timepoint")
plt.xlabel("Signal Intensity")
plt.show()
```

```python
#bar chart
titanic=sns.load_dataset("titanic")
titanic.head(10)
```

```python
sns.barplot(x="class",y="fare",hue="sex",errorbar=None,palette="muted",data=titanic)

plt.xlabel("Class")
plt.ylabel("Fare")

plt.title("Average Fare by Class and Gender on the Titanic")

plt.show()
```

```python
#histogram
iris=sns.load_dataset("iris")
sns.histplot(x="petal_length",data=iris)
```

```python
sns.histplot(x="petal_length",kde=True,bins=20,color="green",data=iris)
plt.xlabel("Petal Length (cm)")
plt.ylabel("Frequency")
plt.title("Distribution of Petal Length in Iris Flowers")

plt.show()
```
