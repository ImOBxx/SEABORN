```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips =sns.load_dataset("tips")

sns.kdeplot(data=tips,x="total_bill")
```

```python
sns.kdeplot(data=tips,x="total_bill",hue="time",fill=True,alpha=0.6,linewidth=2)
plt.title("Density Plot of Total Bill by Meal Time")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
```

```python
#box plotting
sns.boxplot(x="day",y="total_bill",data=tips)
```

```python
#box plotting
sns.boxplot(x="day",y="total_bill",hue="time",palette="Set3",linewidth=2,fliersize=4, data=tips)
plt.title("Box Plot of Total Bill by Meal Time")
plt.xlabel("Day of the week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
#violin plot
iris=sns.load_dataset("iris")
sns.violinplot(x="species",y="petal_length",data=iris)

plt.show()
```

```python
#heatmaps plot

tips=sns.load_dataset("tips")


tips2=tips[["total_bill","tip","size"]]

corr=tips2.corr()

sns.heatmap(corr) 

```

```python
flights = sns.load_dataset('flights')

flights = flights.pivot(index='month',columns='year',values='passengers')

sns.heatmap(flights,cmap="Blues",annot=True,fmt="d")

plt.title("Passengers Per Month")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```

```python
#Pair Plot
sns.pairplot(data=iris)

plt.show()
```

```python
sns.pairplot(data=iris,hue="species",diag_kind="kde",palette="husl")

plt.title("Iris DataSet Pair Plot")

plt.show()
```
